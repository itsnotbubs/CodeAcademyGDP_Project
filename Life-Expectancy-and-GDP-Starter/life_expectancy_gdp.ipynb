```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
```

```python
df = pd.read_csv('all_data.csv')
```

```python
x = df['Year'][0:16]
x
```

```python
plt.figure(figsize=(12,10))
plt.subplot(2,3,1)
plt.title('USA')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.plot(x, df.GDP[df.Country == 'United States of America'])
plt.subplot(2,3,2)
plt.title('Chile')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.plot(x, df.GDP[df.Country == 'Chile'])
plt.subplot(2,3,3)
plt.title('China')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.plot(x, df.GDP[df.Country == 'China'])
plt.subplot(2,3,4)
plt.title('Mexico')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.plot(x, df.GDP[df.Country == 'Mexico'])
plt.subplot(2,3,5)
plt.title('Zimbabwe')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.plot(x, df.GDP[df.Country == 'Zimbabwe'])
plt.subplot(2,3,6)
plt.title('Germany')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.plot(x, df.GDP[df.Country == 'Germany'])
plt.subplots_adjust(wspace=0.3, bottom=0.08)
plt.show()
plt.clf()

```

```python
plt.figure(figsize=(12,10))
plt.subplot(2,3,1)
plt.title('USA')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.plot(x, df['Life expectancy at birth (years)'][df.Country == 'United States of America'])
plt.subplot(2,3,2)
plt.title('Chile')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.plot(x, df['Life expectancy at birth (years)'][df.Country == 'Chile'])
plt.subplot(2,3,3)
plt.title('China')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.plot(x, df['Life expectancy at birth (years)'][df.Country == 'China'])
plt.subplot(2,3,4)
plt.title('Mexico')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.plot(x, df['Life expectancy at birth (years)'][df.Country == 'Mexico'])
plt.subplot(2,3,5)
plt.title('Zimbabwe')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.plot(x, df['Life expectancy at birth (years)'][df.Country == 'Zimbabwe'])
plt.subplot(2,3,6)
plt.title('Germany')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.plot(x, df['Life expectancy at birth (years)'][df.Country == 'Germany'])
plt.subplots_adjust(wspace=0.3, bottom=0.08)
plt.show()
plt.clf()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(x='Country', y='GDP', data=df )
plt.subplots_adjust(wspace=0.3, bottom=0.08)
plt.show()
plt.clf()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(x='Country', y='Life expectancy at birth (years)', data=df )
plt.subplots_adjust(wspace=0.3, bottom=0.08)
plt.show()
plt.clf()
```

```python
sns.scatterplot(x='GDP', y='Life expectancy at birth (years)', hue = 'Country', data=df)
plt.show()
plt.clf()plt
```

```python
corr, p = pearsonr(df.GDP, df['Life expectancy at birth (years)'])
print('There is a ' + str(round(corr, 2) * 100) + '% correlation between GDP and life expectancy')
```
